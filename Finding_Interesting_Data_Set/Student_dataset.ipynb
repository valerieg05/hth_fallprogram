```python
import pandas
import seaborn
import matplotlib.pyplot as plt
import numpy as np

student_data = pandas.read_csv('student_data.csv')
```

```python
student_data.activities.map(dict(yes=1, no=0))
student_data.sex.map(dict(F=1, M=0))
student_data.internet.map(dict(yes=1, no=0))

student_features= student_data[['failures', 'absences', 'sex', 'internet', 'studytime', 'activities']]

corr_heatmap = seaborn.heatmap(student_features.corr(), annot = True, cmap = 'rocket')
plt.show()
```
